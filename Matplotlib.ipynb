#`Matplotlib` es una de las librerías para graficación más populares por ser relativamente sencilla de usar y por su grán versatilidad.

#Vamos a realizar un pequeño ejercicio de graficación usando un montón de puntos aleatorios generados mediante `numpy`, por lo que lo primero será importar ambas librerías de la forma usual:





```
import numpy as np
import matplotlib.pyplot as plt
```

# Ahora formamos dos arreglos de 50 elementos aleatorios. Como ya vimos, ocupamos dar una semilla para luego contruir los arreglos `x` y `y` para las coordenadas.

```
np.random.seed(9601)

N = 50
x = np.random.rand(N)
y = np.random.rand(N)
```

Ya con las x y las y, podemos hacer nuestra primera gráfica con `plt`

```
plt.scatter(x, y)
plt.show()
```

# Listo. Podemos crear algunos atributos para los puntos, generando otros arreglos del mismo tamaño para que allá correspondencia:

```
colors = np.random.rand(N)
area =(30*colors)**2
```

## Se los pasamos a `plt`, diciendole que `s=area` y `c=colors`

```
plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```
